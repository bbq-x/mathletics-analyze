```python
# Populate Odds API Tables

from packages.utils import cnxn_info, season_conversion
import pytz
from datetime import timedelta,datetime
import requests
import pandas as pd

ODDS_API_KEY = os.getenv("ODDS_API_KEY")
SPORT = "baseball_mlb"

odds_dict = {}
for market in ["h2h", "spreads", "totals"]:
    print(market)

    # "Get Odds"
    api_call = f"https://api.the-odds-api.com/v4/sports/{SPORT}/odds/?apiKey={ODDS_API_KEY}&regions=us&markets={market}"
    # api_call = mu.historical_odds_url().format(API_KEY,market,utc_str)  # API Request
    response = requests.get(api_call)                                   # API Response
    json = response.json()                                              # Usable format

    odds_dict[market] = json
    print(response.headers["X-Requests-Last"])
    print(response.headers["X-Requests-Remaining"], "\n")
```

```python
# Event level
for event in odds_dict["totals"]:      # Each element of the response is an event
    id = event["id"]    # Unique id for event
    print(id)
    
    for bookmaker in event["bookmakers"]:       # Each event has bookmakers
        print(bookmaker["key"])                 # Bookmaker name
        print(bookmaker["last_update"])         # Last update of the market
        # now                                   # Add now for time of query, this stamp should match all odds tables
        
        
        print(bookmaker["markets"][0]["outcomes"])  # Always two-item list, one for each team 
        # Convert to df and then add all above data to each line
        df_bm = pd.DataFrame(bookmaker["markets"][0]["outcomes"])
        df_bm["id"] = id
        df_bm["bookmaker"] = bookmaker["key"]
        df_bm["last_update_utc"] = bookmaker["last_update"]
        df_bm["query_time_est"] = now
        print(df_bm)
        
```
